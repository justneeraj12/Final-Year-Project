```python
import pandas as pd

df = pd.read_excel("playground.xlsx")

df.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['b'])
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(16,8))

plt.plot(df['g'].head(1000)) 
plt.plot(np.where(df['g'].head(1000)> 75,df['g'].head(1000), None), color="red", label="1")
plt.plot(np.where(df['g'].head(1000) <60,df['g'].head(1000), None), color="red", label="1")
plt.axhline(60, color='green')
plt.axhline(75, color='green')
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(16,8))

plt.plot(df['p'].head(10000)) 
plt.plot(np.where(df['p'].head(10000)> 0,df['p'].head(10000), None), color="red", label="1")
plt.plot(np.where(df['p'].head(10000) <0,df['p'].head(10000), None), color="blue", label="1")
plt.axhline(0, color='green')
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(16,8))

plt.plot(df['r'].head(1000)) 
plt.plot(np.where(df['r'].head(1000)> 0,df['r'].head(1000), None), color="red", label="1")
plt.plot(np.where(df['r'].head(1000) <0,df['r'].head(1000), None), color="blue", label="1")
plt.axhline(0, color='green')
plt.show()
```
